```python
import requests
import json
from requests.auth import HTTPBasicAuth

a=requests.get('https://api.github.com/orgs/CodingNinjasCodes/repos',auth=('gupta.XXXX26@gmail.com','XXXX'))
pdata=json.loads(a.text)

#print(a.url)
d=pdata
for i in d:
    if i['fork']==False:
        print(i['name'],end=' ')
        print(i['watchers_count'],end=' ')
        print(i['forks_count'])
```
