```python
import pandas as pd
iris=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
df=iris.copy()
df.columns=['sl','sw','pl','pw','Species']

df1=df[df.Species=='Iris-setosa'].describe()
df2=df[df.Species=='Iris-versicolor'].describe()
df3=df[df.Species=='Iris-virginica'].describe()

print(df1)
cols=4
min=3
max=7
avg=1


for i in range(cols):
    
    print("{0:.2f}".format(df1.iloc[min,i]),end=' ')
print('iris-setosa')
print()
for i in range(cols):
    print("{0:.2f}".format(df1.iloc[max,i]),end=' ')
print('iris-setosa')
print()
for i in range(cols):
    print("{0:.2f}".format(df1.iloc[avg,i]),end=' ')
print('iris-setosa')
print()

for i in range(cols):
    print("{0:.2f}".format(df2.iloc[min,i]),end=' ')
print('Iris-versicolor')
print()
for i in range(cols):
    print("{0:.2f}".format(df2.iloc[max,i]),end=' ')
print('Iris-versicolor')
print()
for i in range(cols):
    print("{0:.2f}".format(df2.iloc[avg,i]),end=' ')
print('Iris-versicolor')
print()

for i in range(cols):
    print("{0:.2f}".format(df3.iloc[min,i]),end=' ')
print('Iris-virginica')
print()
for i in range(cols):
    print("{0:.2f}".format(df3.iloc[max,i]),end=' ')
print('Iris-virginica')

for i in range(cols):
    print("{0:.2f}".format(df3.iloc[avg,i]),end=' ')
print('Iris-virginica')

    
```

```python
import pandas as pd
iris=pd.read_csv("D:\\Dropbox\\terrorismData.csv")
df=iris.copy()
df=df[(df.Country=='India') & (df.Day!=0) & (df.Month!=0)]

df['Date']=pd.to_datetime(df[['Year', 'Month', 'Day']])

df=df[(df.Date>='2014-05-26')]
print(df.shape[0],end=' ')
d={}
for i in df.Group:
    if(i!='Unknown'):
        d[i]=d.get(i,0)+1

max=0
for i in d:
    if d[i]>max:
        max=d[i]
        temp=i
print(temp)
```

```python
import pandas as pd
iris=pd.read_csv("D:\\Dropbox\\terrorismData.csv")
df=iris.copy()
df.Killed.fillna(0,inplace=True)
df.Wounded.fillna(0,inplace=True)
df['Casuality']=df['Killed']+df['Wounded']
df1=df[df.State=='Jammu and Kashmir']

freq1=df1.Casuality.sum()//df1.Year.nunique()

df2=df[(df.State=='Jharkhand') | (df.State=='Odisha') | (df.State=='Andhra Pradesh') | (df.State=='Chhattisgarh')]
freq2=df2.Casuality.sum()//df2.Year.nunique()
print(freq2,end=" ")
print(freq1)
```
