```python
import matplotlib.pyplot as pt
x1=[0,1,2,3,4,5,6,7,9,10]
y1=[1,2,5,10,17,26,37,50,82,101]
x2=[i for i in range(0,40,2)]
y2=[]
for i in x2:
    y2.append(i)
pt.xlabel("x")
pt.ylabel("y")
pt.scatter(x1,y1)
pt.plot(x1,y1,label="x1y1")

pt.scatter(x2,y2)
pt.plot(x2,y2,label="x2y2")
pt.legend()
pt.grid()
pt.show()
```
