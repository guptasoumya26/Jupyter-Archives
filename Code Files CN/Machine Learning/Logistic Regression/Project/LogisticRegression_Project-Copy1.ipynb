# Multiclass Logistic Regression for Air Plane Crash Survival Prediction

## Importing Training Data for Model

```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
import pandas as pd


#Loading Training Data
df=pd.read_csv("training_titanic_x_y_train.csv")

```

## Replacing NA values with Mean in Age Column

```python
#Replacing NA with Mean
df.Age.fillna(value=df.Age.mean(),inplace=True)
df.head()
```

## Dropping Irrelevant cols for Plane Crash prediction

```python
#Dropping irrelevant cols
del df['Name']
del df['Ticket']
del df['Cabin']
del df['Fare']
df.head()
```

## Fixing Categorical columns via One Hot Encoder

```python
#Fixing Sex and Embarked cols
df=pd.get_dummies(df,columns=["Sex","Embarked"])
df.head()
```

## Making Training Data for model fitting

```python
#Getting X and Y for training
X_train=df[['Pclass','Age','SibSp','Parch','Sex_female','Embarked_C','Embarked_Q']].values
y_train=df['Survived']
```

## Training the model

```python
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)

#Training the Model
model=LogisticRegression()
model.fit(X_train,y_train)
```

## Calculating model score on training data itself

```python
#Calculating Model score after training

model.score(X_train,y_train)
```

##  Importing Testing Data

```python
#Loading test data set
df1=pd.read_csv("test_titanic_x_test.csv")
```

## Replacing NAN values in Age Column

```python
#Replacing NA with Mean
df1.Age.fillna(value=df.Age.mean(),inplace=True)
df1.head()
```

## Dropping irrelevant cols with regards to plane crash

```python
#Dropping irrelevant cols
del df1['Name']
del df1['Ticket']
del df1['Cabin']
del df1['Fare']
df1.head()
```

## Fixing categorical values via One Hot Encoder

```python
#Fixing Sex and Embarked cols
df1=pd.get_dummies(df1,columns=["Sex","Embarked"])
df.head()
```

## Preparing testing data

```python
#Getting X of testing
X_test=df1[['Pclass','Age','SibSp','Parch','Sex_female','Embarked_C','Embarked_Q']].values
```

## Saving the predictions in a CSV File , predictions.csv

```python
pred=model.predict(X_test)
#Saving predictions
l=[]
for i in pred:
    
    l.append(int(i))
y_test_pred=np.array(l)

np.savetxt("predictions2.csv",y_test_pred,delimiter=",",fmt='%d')
```
