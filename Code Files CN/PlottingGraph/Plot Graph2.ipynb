```python
import pandas as pd
import matplotlib.pyplot as pt
iris=pd.read_csv("D:\\Dropbox\\amazon_jobs_dataset.csv")
df=iris.copy()
rows=df.location.shape[0]
months=[]
#print(df.iloc[i,3].split())
for i in range(rows):
    month=df.iloc[i,3].split()[0]
    
    months.append(month)

d={}
for i in months:
    d[i]=d.get(i,0)+1

x=list(d.keys())


y=list(d.values())
pt.scatter(x,y)
pt.xticks(rotation=40)
pt.plot(x,y)
pt.show()

for i in d:
    print(i,end=' ')
    print(d[i])
```

```python
import pandas as pd
import matplotlib.pyplot as pt
iris=pd.read_csv("D:\\Dropbox\\amazon_jobs_dataset.csv")
df=iris.copy()
d={}
for i in range(df.shape[0]):
    if 'IN' in df.iloc[i,2].split(",")[0]:
        d[df.iloc[i,2]]=d.get(df.iloc[i,2],0)+1

for i in d:
    print(i.split(",")[2],end=' ')
    print(d[i])
labels=["Pune","Gurgaon","Chennai","Hyderbad","Bangalore"]
sizes=[1,5,33,36,66]
pt.pie(sizes,labels=labels,autopct="%.2f%%")
pt.axis("equal")
pt.show()
```
