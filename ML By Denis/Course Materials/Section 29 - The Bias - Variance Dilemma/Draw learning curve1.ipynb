## learning curve

```python
import pandas as pd

df = pd.read_csv("./classification.csv")

df.head()
```

```python
X = df[["age", "interest"]].values
y = df["success"].values
```

```python
# Not actually needed - but sets a few
# parked Matplotlib properties to make the graphic
# is generated in a higher resolution
import helper
```

```python
from sklearn.model_selection import learning_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle

import numpy as np

X, y = shuffle(X, y)

train_sizes_abs, train_scores, test_scores = learning_curve(
    LogisticRegression(C = 100000000000000), X, y, cv = RepeatedStratifiedKFold(n_repeats = 1000))
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.plot(train_sizes_abs, np.median(train_scores, axis = 1))
plt.plot(train_sizes_abs, np.median(test_scores, axis = 1))

plt.show()
```
