```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(4)
x = x**2
x2 = x**3
print(x)
```

```python
plt.plot(x)
plt.plot(x2,color="red")
plt.show()
```

```python
plt.scatter(x,x*2,marker='^',color='orange',label="Stock Price")
plt.legend()
plt.xlabel("days")
plt.ylabel("price")
plt.title("STock Price Prediction")
plt.show()
```
