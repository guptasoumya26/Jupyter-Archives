```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)

styles = plt.style.available
plt.style.use(styles[4])

Y = Y.reshape((-1,1))
```

```python
#Normalize the Data
X = (X - X.mean())/X.std()
plt.figure(0)
plt.scatter(X,Y)
plt.title("Normalized Data Plot")
plt.show()

```

```python
ones = np.ones((X.shape[0],1))
print(X.shape)
print(ones.shape)
X_ = np.hstack((X,ones))
print(X_.shape)
```

```python
# Hypothesis Same for both linear and locally weighted regression
def getHypothesis(X,theta):
    return np.dot(X,theta)

#Linear Regression, Predict for a Query Point
def getTheta(X,Y):  
    
    #Using the formula : theta =(X′X)inv * X′Y
    Y = np.mat(Y)
    
    firstPart = np.dot(X.T,X)
    secondPart = np.dot(X.T,Y)
    theta = np.linalg.pinv(firstPart)*secondPart
    return theta
```

```python
theta = getTheta(X_,Y)
print(theta)
```

```python
Ypredicted  = getHypothesis(X_,theta)
Ypredicted.shape
```
